```python
import pandas as pd
```

## CAFE

```python
def prepare_single_cafe_df(df_cafe):
    # Ganti nama kolom 'cafe: (Indonesia)'
    if 'cafe: (Indonesia)' in df_cafe.columns:
        df_cafe.rename(columns={'cafe: (Indonesia)': 'cafe_value'}, inplace=True)
    elif df_cafe.shape[1] > 1 and df_cafe.columns[1]: # Fallback if exact name not found but it's the second column
        df_cafe.rename(columns={df_cafe.columns[1]: 'cafe_value'}, inplace=True)

    # Konversi kolom 'Day' ke datetime
    df_cafe['Day'] = pd.to_datetime(df_cafe['Day'])
    return df_cafe
```

```python
ex_cafe_1_df = pd.read_csv("external_data/cafe-1.csv", skiprows=1)
ex_cafe_2_df = pd.read_csv("external_data/cafe-2.csv", skiprows=1)
ex_cafe_3_df = pd.read_csv("external_data/cafe-3.csv", skiprows=1)

ex_cafe_1_df = prepare_single_cafe_df(ex_cafe_1_df)
ex_cafe_2_df = prepare_single_cafe_df(ex_cafe_2_df)
ex_cafe_3_df = prepare_single_cafe_df(ex_cafe_3_df)

ex_cafe_df = pd.concat([ex_cafe_1_df, ex_cafe_2_df, ex_cafe_3_df], ignore_index=True)
ex_cafe_df = ex_cafe_df.set_index('Day')
ex_cafe_df.index.name = 'date'
ex_cafe_df = ex_cafe_df.sort_index()

ex_cafe_df
```

```python
def prepare_external_df(df_ext, date_col_name='date'):
    if date_col_name not in df_ext.columns:
        # If the date column is not named 'date', you'll need to adjust here
        # E.g., if it's named 'time'
        if 'time' in df_ext.columns:
            date_col_name = 'time'
        else:
            print(f"Warning: Date column '{date_col_name}' or 'time' not found in DataFrame. Skipping preparation.")
            return None # Or handle error appropriately

    df_ext[date_col_name] = pd.to_datetime(df_ext[date_col_name])
    df_ext = df_ext.set_index(date_col_name)
    df_ext.index.name = 'date' # Standardize index name to 'date' for merging
    return df_ext
```
